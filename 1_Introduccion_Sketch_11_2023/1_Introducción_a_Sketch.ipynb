```python
# Get the earliest date in the 'FECHA' column
earliest_date = df['FECHA'].min()

# Print the result
print('La fecha en la que se realiz la venta ms temprana fue:', earliest_date)
```

```python
df.sketch.ask("cual es el producto mas vendido")
```

```python
df.sketch.howto("cual es el producto mas vendido")
```

```python
# Get the product with the highest total quantity sold
most_sold_product = df.groupby('PRODUCTO')['CANTIDAD'].sum().sort_values(ascending=False).index[0]

# Print the result
print('El producto ms vendido es:', most_sold_product)
```

```python
df.sketch.ask("cuales son los productos con baja demanda")
```

```python
df.sketch.howto("cuales son los productos con baja demanda?")
```

```python
# Get the counts of each product
product_counts = df.groupby('PRODUCTO')['CANTIDAD'].sum()

# Get the products with low demand
low_demand_products = product_counts[product_counts < 3]

# Print the list of low demand products
print(low_demand_products.index.tolist())
```

```python

import matplotlib.pyplot as plt


# Plot a boxplot of the 'PRECIO' column
df.boxplot(column='PRECIO')

```

```python
df.sketch.howto("realiza una serie de tiempo para ver el comportamiento de las ventas por da")
```

```python
# Create a time series of sales by day
df_ts = df.groupby('FECHA')['CANTIDAD'].sum().reset_index()

# Plot the time series
plt.plot(df_ts['FECHA'], df_ts['CANTIDAD'])
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Sales by Day')
plt.show()
```
